```python
import numpy as np
from PIL import Image, ImageFont, ImageDraw
from PIL.ImageChops import add, subtract, multiply, difference, screen
import PIL.ImageStat as stat
from skimage.io import imread, imsave, imshow, show
from skimage import color, viewer, img_as_float, data
from skimage.transform import SimilarityTransform, warp, swirl
from skimage.util import random_noise
import matplotlib.image as mpimg
import matplotlib.pylab as plt
from scipy import misc
```

# 다양한 영상 유형과 파일 형식 다루기 및 기본 영상 조작

```python
im = Image.open("../images/parrot.png") 	# PIL 함수 이용 영상 읽기
print(im.mode) 								# RGB
im.save("../images/parrot.jpg") 			# Image 객체 내의 save() 함수로 영상 저장
```

```python
im = Image.open("../images/hill.png") 			# 알파 채널 포함 영상
print(im.mode) 									# RGBA
im.convert('RGB').save("../images/hill.jpg")	# 알파 채널 있는 경우 변환 필요
```

```python
한 영상 모드에서 다른 영상 모드로 변환
```

```python
im = imread("../images/parrot.png", as_gray=True)
print(im.shape) 			#(362, 486)
```

```python
im = imread("../images/Ishihara.png")
im_g = color.rgb2gray(im)
plt.subplot(121), plt.imshow(im, cmap='gray'), plt.axis('off')
plt.subplot(122), plt.imshow(im_g, cmap='gray'), plt.axis('off')
plt.show()
```

```python
한 색상 공간에서 다른 색상 공간으로 변환
```

```python
im = imread("../images/parrot.png")
im_hsv = color.rgb2hsv(im)
plt.gray() # 명암도 영상으로 표시

plt.figure(figsize=(10,8))
plt.subplot(221), plt.imshow(im_hsv[...,0])
plt.title('h', size=20), plt.axis('off')
plt.subplot(222), plt.imshow(im_hsv[...,1]),
plt.title('s', size=20), plt.axis('off')
plt.subplot(223), plt.imshow(im_hsv[...,2]),
plt.title('v', size=20), plt.axis('off')
plt.subplot(224), plt.axis('off')
plt.show()
```

```python
영상 데이터 구조 변환
```

```python
im = Image.open('../images/flowers.png') # PIL 모듈의 Image 객체로 영상 읽기
im = np.array(im)						 # Image 객체에서 numpy ndarray 생성
imshow(im)					 # skimage 모듈의 imshow 함수 사용해 영상 표시
plt.axis('off')
show()
```

```python
PIL Image 객체로 변환
```

```python
im = imread('../images/flowers.png') # numpy ndarray로 영상읽기
im = Image.fromarray(im) # PIL 영상 객체 생성
im.show() # PIL Image.show 함수 사용해 영상표시
```
